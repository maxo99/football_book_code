```python
import pandas as pd
import nfl_data_py as nfl
import os
from loader import get_pbp

## Load data from the package
pbp_py = get_pbp(2021)

print(f"shape:{pbp_py.shape}")
# print(f"columns:{pbp_py.columns}")
# pbp_py.describe()
# print(f"isna:{pbp_py.isna().sum()}")
# print(f"dtypes:{pbp_py.dtypes}")
pbp_py.sample(10)
```

```python

## filter out passing data
pbp_py_p = (
    pbp_py.query('play_type == "pass" & air_yards.notnull()')
    .groupby(["passer_id", "passer"])
    .agg({"air_yards": ["count", "mean","max"]})
)
pbp_py_p.sample(10)
```

```python
## format and print data for passing plays
pbp_py_p.columns = list(map("_".join, pbp_py_p.columns.values))

pbp_py_p.query("air_yards_count > 100").sort_values(
    by="air_yards_mean", ascending=[False]
)
```
